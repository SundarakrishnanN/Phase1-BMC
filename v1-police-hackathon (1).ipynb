```python
from ultralytics import YOLO
import os
import cv2

from PIL import Image
```

```python
os.environ['WANDB_DISABLED'] = 'true'
```

```python

dir='/kaggle/input/v3-police-data/v3_total_data_combined/data.yaml'
model = YOLO("yolov10n.pt")

results = model.train(data=dir, epochs=300,dropout=0.2,label_smoothing=0.3,patience=50)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
image_dir = "/kaggle/input/pls-delete-soon"
image_files = os.listdir(image_dir)


num_images_to_display = 1
for i in range(num_images_to_display):

    img_path = os.path.join(image_dir, image_files[i])
    img = cv2.imread(img_path)
    results=model(img)
    r=results[0]
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
        ##im.show()  # show image
    plt.imshow(im)
    plt.show()
```

```python
import cv2
from ultralytics import YOLO
from IPython.display import display, Image, clear_output

# Load the YOLO model
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")  # You can replace with the path to your YOLO model

# Open the video file
cap = cv2.VideoCapture('/kaggle/input/v2-data-police/Devasandra_Sgnl_JN_FIX_1_time_2024-05-14T07_30_02_001.mp4')  # Replace with your video file path

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Perform object detection
    results = model(frame)

    # Iterate over results and render each one
    for result in results:
        annotated_frame = result.plot()

    # Display the annotated frame
    _, encoded_img = cv2.imencode('.jpg', annotated_frame)
    display(Image(data=encoded_img.tobytes()))
    clear_output(wait=True)



# Release the video capture object
cap.release()
```
